```python
import pandas as pd

df = pd.read_csv('Auckland.csv', skiprows=18)
df['Date'] = pd.to_datetime(df['Date(NZST)'], format='%Y%m%d')

print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(ncols=1, nrows=4, constrained_layout=True, figsize=(15, 8))
sns.lineplot(x='Date', y='Pmsl(hPa)', data=df[df['Station'] == 'Auckland Aero'], ax=axes[0])
axes[0].set_title('Auckland')
sns.lineplot(x='Date', y='Pmsl(hPa)', data=df[df['Station'] == 'Hamilton Aws'], ax=axes[1])
axes[1].set_title('Hamilton')
sns.lineplot(x='Date', y='Pmsl(hPa)', data=df[df['Station'] == 'Paeroa Aws'], ax=axes[2])
axes[2].set_title('Paeroa')
sns.lineplot(x='Date', y='Pmsl(hPa)', data=df[df['Station'] == 'Westport Aero Aws'], ax=axes[3])
axes[3].set_title('Westport')
```
