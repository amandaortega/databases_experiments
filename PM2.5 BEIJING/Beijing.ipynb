```python
import pandas as pd

df = pd.read_csv('PRSA_data_2010.1.1-2014.12.31.csv')
print(df.head())
```

```python
df_ = df[((df['year'] == 2014) & (df['month'] == 11) & (df['day'] >= 22)) | ((df['year'] == 2014) & (df['month'] == 12) & (df['day'] <= 31))]

print(df_)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

fig = plt.figure(1, figsize=(15, 2))
fig.add_subplot(111)
sns.lineplot(y='pm2.5', x=np.arange(df_.shape[0]), data=df_)
```
