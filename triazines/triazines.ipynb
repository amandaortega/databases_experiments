```python
sns.relplot(x='index', y='activity', kind='line', data=dataset.reset_index())
```

```python
dataset.isna().sum().sum()
```

```python
train_dataset = dataset.sample(n=100,random_state=10)
test_dataset = dataset.drop(train_dataset.index)
```

```python
train_labels = train_dataset.pop('activity')
test_labels = test_dataset.pop('activity')
```

```python
import os

def save(path):
    if not os.path.exists(path):
        os.makedirs(path)

    np.savetxt(path + 'X_train.csv', train_dataset)
    np.savetxt(path + 'X_test.csv', test_dataset)
    np.savetxt(path + 'Y_train.csv', train_labels)
    np.savetxt(path + 'Y_test.csv', test_labels)    

path = '/home/amanda/Dropbox/trabalho/testes/aplicacoes/triazines/data/'
save(path)

path = '/home/amanda/Dropbox/trabalho/codigos/propostas/EVeP/data/triazines/'
save(path)
```
